```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
from scipy.stats import pearsonr
import math
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import ttest_ind
from scipy import stats
```

```python
data = pd.read_csv("C:Seasons_Stats.csv")
df = pd.DataFrame(data)
#df.describe()
#df.Pos.head(20)
```

```python
#preprocesing of data with all conditions
#dfe is preprocesing list with all attributes
dfe=df[(df.Year>1980 ) & (df.Age>18)& (df.Age<40)& (df.MP>500) & (df.Pos.isin(['PG','SG','SF','PF','C']))]
dfe.describe()
#dfe.isnull().values.any()
```

```python
#keep only important columns
players=dfe.filter(items=['Year','Player','Pos','Age','Tm','G','PTS','AST','TRB','BLK','PER','TS%'])
#players.describe()
#players.info()
players.isnull().values.any()
players=players.dropna(subset=['Year','Player','Pos','Age','Tm','G','PTS','AST','TRB','BLK','PER','TS%' ])
#players.head(20)
```

```python
#erwtima A

#oi synolikoi pontoi ana paixti
sumPointsPp=players.groupby(['Player'], sort=False ,dropna=False)['PTS'].sum()

#kata fthinoysa seira tajinomisi
n=sumPointsPp.to_numpy()
m=-np.sort(-n)

#join with the dataframe players to have all in one
player_total_Points = pd.merge(sumPointsPp, players, on=['Player'], how='inner')
```

```python

# erwtima B
#kathe paixtis se osa paixnidia kai na paije se mia sezon therwetai ws ena atomo
#gia tous pontous
playersB1=players
playersB1=players.groupby(['Age','Player'], sort=False ,dropna=False,as_index=False)[['PTS']].sum()

#sumof G for gia kathe paixti
df_sG=players.groupby(['Age','Player'], sort=False ,dropna=False,as_index=False)[['G']].sum()

def kane_mean_points(df):
    for i in df.index:
        df.at[i,"PPG"] =df.at[i,"PTS"]/df_sG.at[i,"G"]

#df_sG
kane_mean_points(playersB1)
#playersB1

#gia assist
playersB2=players
playersB2=players.groupby(['Age','Player'], sort=False ,dropna=False,as_index=False)[['AST']].sum()
def kane_mean_ast(df):
    for i in df.index:
        df.at[i,"APG"] =df.at[i,"AST"]/df_sG.at[i,"G"]

kane_mean_ast(playersB2)
playersB2

#gia rebount
playersB3=players
playersB3=players.groupby(['Age','Player'], sort=False ,dropna=False,as_index=False)[['TRB']].sum()
def kane_mean_r(df):
    for i in df.index:
        df.at[i,"RPG"] =df.at[i,"TRB"]/df_sG.at[i,"G"]

kane_mean_r(playersB3)
playersB3.head(20)

#gia block

playersB4=players
playersB4=players.groupby(['Age','Player'], sort=False ,dropna=False,as_index=False)[['BLK']].sum()
def kane_mean_b(df):
    for i in df.index:
        df.at[i,"BPG"] =df.at[i,"BLK"]/df_sG.at[i,"G"]

kane_mean_b(playersB4)
#playersB4.head(20)
```

```python

dfstat=dfstat[dfstat.PLAYER_ID==201566]

dfstat=dfstat.dropna(subset=['PLAYER_ID','PTS','AST','REB'])

all_games=dfstat['GAME_ID'].count()
all_games

#na ferei triple double PTS 
dftd=dfstat[(dfstat.PTS>=10 ) & (dfstat.AST>=10 )& (dfstat.REB>=10 )]
dftd
d=dftd['GAME_ID'].count()

#pithanotita triple douple 
ptrip=d/all_games
ptrip

```

```python
#gia na brw posa paixnidia exei kerdisei h omada poy paizei exontas knaei triple kanw join

jointrip=pd.merge(dftd, dfwins, on=['GAME_ID'], how='inner')#yparxoyn 2 egrafes parapanw mporei mia omada na grafthke dyo fores ws host kai os visitor
#jointrip.drop_duplicates()
#jointrip.head()
```

```python

#keep only important columns
jointrip1=jointrip.filter(items=['GAME_ID','TEAM_ID_home','TEAM_ID','HOME_TEAM_WINS'])
#checkhost(jointrip)
#home id == team id ara win mas deixnei an kerdise
jointrip1=jointrip[jointrip.TEAM_ID_home==jointrip.TEAM_ID]
#metraw toys assoys
jointrip1=jointrip1[jointrip1.HOME_TEAM_WINS==1]
l1win=len(jointrip1)
#home id =! team id ara win DEN mas deixnei an kerdise
jointrip2=jointrip[jointrip.TEAM_ID_home!=jointrip.TEAM_ID]
jointrip2=jointrip2[jointrip2.HOME_TEAM_WINS==0]
l2win=len(jointrip2)
lallwin=l1win+l2win
lallwin
ptomis=lallwin/all_games
ptomis
```
